```python
#Importing the pandas library
import pandas as pd
#Using the 'read_csv' of the pandas library to read the dataset
df = pd.read_csv('Data_Science_Fields_Salary_Categorization.csv')
```

```python
#Printing the dataset to observe the columns
df
```

```python
#Observing the datatypes of the columns in the dataframe and looking for the Null values
df.info()
```

```python
#Cleaning and Data Preprocessing

#As observed above the Salry_In_Rupees column is an object and it does have some ',' in it
#We need to eleminate those to further convert it into a numeric format

df['Salary_In_Rupees'] = df['Salary_In_Rupees'].apply(lambda i:i.replace(',',''))
```

```python
#Checking if the values in salary column are properly structured after applying of the replace function
df.head(10)
```

```python
#Salary in rupees column is in object format so first converting it into a numeric format

df['Salary_In_Rupees'] = pd.to_numeric(df['Salary_In_Rupees'])

#checking if the #Checking if the datatype is converted
df.info()
```

```python
#This library is used to convert the COUNTRY CODES that are in ISO 3166 format to the Country names
import pycountry
```

```python
'''Creating two lists, one to store the Country name of the country code of employee location in the dataframe, and the other 
to store the country name of the country code of company location in the dataframe.'''
emp_country_list = []
comp_country_list = []
#For loop to iterate through the datfarame 'Employee_Location column'
for country_code in df.Employee_Location:
    #For each country code in the Employee Location column appending the respective country name
    #This is storing each Country name in same order as in dataframe
    emp_country_list.append(pycountry.countries.get(alpha_2=country_code).name)

for country_code in df.Company_Location:
    #For each country code in the Company Location column appending the respective country name
    #This is storing each Country name in same order as in dataframe
    comp_country_list.append(pycountry.countries.get(alpha_2=country_code).name)

#Updating the 'Employee_Location' and 'Company_Location' columns
df['Employee_Location'] = emp_country_list
df['Company_Location'] = comp_country_list
```

```python
#Checking if the dataframe was successfully updated
df.head()
```

```python
'''Question 1- If company location and employee location is similar then is the salary higher than the average or 
lower thanÂ theÂ average?'''

#Creating a Location_Match column and setting all the values to 'Similar'
df['Location_Match'] = 'Similar'
#At each location if the 'Employee_Location' and 'Company_Location' are not the same then set 'Location_Match' to 'Not Similar'.
df.loc[df['Employee_Location'] != df['Company_Location'], 'Location_Match'] = 'Not Similar'
```

```python
#Creating a new_df to only get those rows whose 'Location_Match' column has value is 'Similar'
df_similar = df[df['Location_Match'] == 'Similar']
#Now creating a new column named 'Salry_grater_than_mean' to find that if the salary is grater than mean or not
#For this, first setting all the values in this new column to 'Yes'
df_similar['Salary_greater_than_mean'] = 'Yes'
#Setting the values of 'Salary_greater_than_mean' to 'No' if the value in 'SaLary_In_USD' is less than the mean salary
df_similar.loc[df_similar['Salary_In_USD'] < df['Salary_In_USD'].mean(), 'Salary_greater_than_mean'] = 'No'
```

```python
#Checking the df_similar
df_similar.head(10)
```

```python
#Using groupby on 'Salry_greater_than_mean' column and counting the number or occurences of 'Yes' and 'No' using 'count()'
final_answer_1 = df_similar.groupby('Salary_greater_than_mean')['Salary_greater_than_mean'].count()
final_answer_1
```

```python
#Converting the final output to a dataframe
final_answer_1 = final_answer_1.to_frame()
```

```python
#Saving the final answer to the question in a excel file
final_answer_1.to_excel("output_question_1.xlsx", index=True)
```

```python
#Question 4 - What is the salary trend according to company size?
```

```python
#Creting three new dataframes to store the data of three sizes of companies Mid-scale, Large-scale, Small-scale
df_M = df[df['Company_Size']=='M']
df_L = df[df['Company_Size']=='L']
df_S = df[df['Company_Size']=='S']
```

```python
#Finding average, minimum, and maximum salary of Small-scale companies
average_salary_of_small_company = df_S['Salary_In_USD'].mean()
min_salary_of_small_company = df_S['Salary_In_USD'].min()
max_salary_of_small_company = df_S['Salary_In_USD'].max()
```

```python
#Finding average, minimum, and maximum salary of Medium-scale companies
average_salary_of_medium_company = df_M['Salary_In_USD'].mean()
min_salary_of_medium_company = df_M['Salary_In_USD'].min()
max_salary_of_medium_company = df_M['Salary_In_USD'].max()
```

```python
#Finding average, minimum, and maximum salary of Large-scale companies
average_salary_of_large_company = df_L['Salary_In_USD'].mean()
min_salary_of_large_company = df_L['Salary_In_USD'].min()
max_salary_of_large_company = df_L['Salary_In_USD'].max()
```

```python
#Creating a dictionary so that it is easier to convert to a dataframe
overall_dict = {
    'max':[max_salary_of_small_company, max_salary_of_medium_company, max_salary_of_large_company],
    'min':[min_salary_of_small_company, min_salary_of_medium_company, min_salary_of_large_company],
    'average':[average_salary_of_small_company, average_salary_of_medium_company, average_salary_of_large_company],
    'company_size':['small','medium','large']
}
```

```python
#Converting the dictionary to dataframe
final_answer_4 = pd.DataFrame.from_dict(overall_dict)
final_answer_4
```

```python
#Saving the answer to question 4 in a excel file
final_answer_4.to_excel("output_question_4.xlsx", index=True)
```
